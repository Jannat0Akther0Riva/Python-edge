```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load the dataset
from google.colab import files
  # Upload the 'train.csv' file

data = pd.read_csv('train.csv')
```

```python
# Task 1: Find the total fare paid by passengers in each class
fare_per_class = data.groupby('Pclass')['Fare'].sum()
print("Total fare paid by passengers in each class:")
print(fare_per_class)
```

```python
# Task 2: Plot a pie chart showing the proportion of passengers who traveled alone
# Passengers are considered "traveling alone" if SibSp and Parch are both 0.
data['TravelAlone'] = (data['SibSp'] + data['Parch']) == 0
alone_counts = data['TravelAlone'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(alone_counts, labels=['Alone', 'With Family'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title("Proportion of Passengers Traveling Alone")
plt.show()
```

```python
# Task 3: Determine the survival rate for passengers with no family members onboard
survival_rate_alone = data[data['TravelAlone']]['Survived'].mean()
print(f"Survival rate for passengers traveling alone: {survival_rate_alone:.2f}")
```

```python
# Task 4: Create a swarm plot showing survival based on age
plt.figure(figsize=(10, 6))
sns.swarmplot(data=data, x='Survived', y='Age', palette='Set2')
plt.title("Swarm Plot of Survival Based on Age")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
# Task 5: Count the number of children who survived
# Define children as passengers under 18 years old.
children_survived = data[(data['Age'] < 18) & (data['Survived'] == 1)].shape[0]
print(f"Number of children who survived: {children_survived}")
```

```python


# Task 6: Analyze the distribution of ages among survivors using a KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data[data['Survived'] == 1], x='Age', shade=True, color='green', label='Survived')
sns.kdeplot(data=data[data['Survived'] == 0], x='Age', shade=True, color='red', label='Did Not Survive')
plt.title("Age Distribution among Survivors and Non-Survivors")
plt.xlabel("Age")
plt.ylabel("Density")
plt.legend()
plt.show()
```
